```python
import pandas as pd
import numpy as np
import seaborn as sns
import sweetviz as sv
```

```python
import csv
def load_csv(filename):
    """Helper to load the tournament data from csv."""
    rows = []
    with open(filename) as f:
        reader = csv.DictReader(f)
        rows.extend(iter(reader))
    return rows
```

```python
fp_one = '/Users/christopherfiore/Documents/GitHub/cwl-data/data/data-2017-08-13-champs.csv'
fp_two = '/Users/christopherfiore/Documents/GitHub/cwl-data/data/data-2018-08-19-champs.csv'
champs_2017 = pd.DataFrame(load_csv(fp_one))
champs_2018 = pd.DataFrame(load_csv(fp_two))


```
